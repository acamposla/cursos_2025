# **stateful** y **stateless** 





### 1. Â¿QuÃ© significa "stateful"?

- **Stateful = "con memoria"**
- Imagina que tienes una pizarra en la que cada vez que dibujas, el programa recuerda **automÃ¡ticamente** dÃ³nde estÃ¡s dibujando.
- **pyplot** es una interfaz stateful:  
  - Cuando llamas a funciones como `plt.plot()`, Matplotlib ya tiene un "lienzo" (la figura y los ejes) **creados y listos** para usarse.
  - **No necesitas decirle** en quÃ© figura o eje dibujar: Matplotlib lo hace por ti.
- **Ventaja:** Es rÃ¡pido y fÃ¡cil para hacer grÃ¡ficos simples en una sesiÃ³n interactiva.
- **Ejemplo sencillo:**
  ```python
  import matplotlib.pyplot as plt
  import numpy as np

  x = np.arange(0, 1, 0.1)
  y = np.arange(0, 10, 1)
  plt.plot(x, y)  # Dibuja la lÃ­nea en la "pizarra" que Matplotlib ya tiene preparada.
  plt.title("GrÃ¡fico sencillo")
  plt.show()      # Muestra el grÃ¡fico.
  ```

---

### 2. Â¿QuÃ© significa "stateless" o "orientado a objetos"?

- **Stateless = "sin memoria oculta"**
- En este caso, **tÃº creas y manejas** la pizarra (la figura) y cada parte de ella (los ejes) **manualmente**.
- Esto se conoce como el **enfoque orientado a objetos**.
- **Ventaja:**  
  - Te da **total control** sobre cada parte del grÃ¡fico.
  - Es ideal para crear grÃ¡ficos complejos o con **varios subgrÃ¡ficos** (mÃ¡s de un eje, cada uno con sus propios detalles).
- **Ejemplo sencillo:**
  ```python
  import matplotlib.pyplot as plt
  import numpy as np

  # Creas explÃ­citamente la figura y los ejes
  fig, ax = plt.subplots()
  
  x = np.arange(0, 1, 0.1)
  y = np.arange(0, 10, 1)
  
  ax.plot(x, y)         # Dibuja la lÃ­nea en el eje que tÃº controlas.
  ax.set_title("GrÃ¡fico con control total")
  
  plt.show()            # Muestra el grÃ¡fico.
  ```

---

### 3. Â¿Por quÃ© es importante esta diferencia?

- **Usar solo pyplot (stateful)** es genial para cosas rÃ¡pidas y simples, pero:
  - **Se vuelve confuso** cuando necesitas crear **varios grÃ¡ficos** o modificar partes muy especÃ­ficas de un grÃ¡fico.
- **Usar el enfoque orientado a objetos (stateless)** te permite:
  - **Saber exactamente** en quÃ© figura y en quÃ© eje estÃ¡s trabajando.
  - **Modificar detalles** (como el color de un solo tick, leyendas, etiquetas, etc.) sin que se mezclen con otros grÃ¡ficos.

---

### Resumen Final

- **Stateful (pyplot):**  
  - **AutomÃ¡tico y sencillo.**  
  - El programa "recuerda" la figura y los ejes para ti.
  
- **Stateless (orientado a objetos):**  
  - **Control total y explÃ­cito.**  
  - TÃº creas y manejas cada objeto (figura y ejes) manualmente.

En resumen, para grÃ¡ficos rÃ¡pidos y simples puedes usar **pyplot** (stateful), pero si tu grÃ¡fico se complica o necesitas mucha personalizaciÃ³n, es mejor usar el enfoque orientado a objetos (stateless) para tener un control mÃ¡s claro y preciso.

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(444)

rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng


fig, ax = plt.subplots(figsize=(5, 3))

ax.stackplot(yrs, rng+rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time')
ax.legend(loc='upper left')
ax.set_ylabel('Total debt')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])

fig.tight_layout()
plt.show()
````
